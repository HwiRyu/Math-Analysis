```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import sys
from collections import deque
from IPython.display import display, clear_output
import time

def print_over(text):
    sys.stdout.write('\r' + text)
    sys.stdout.flush()
    
def generate_formatted_string(number): 
    number = f"{number:.14f}"
    int_part, decimal_part = str(number).split('.')
    decimal_part = decimal_part[:13]
    if len(int_part) > 3:
        format_string = 'OUT!OFVALUEPLEASE'
    else:
        num_zeros_needed = max(0, 3 - len(int_part))
        int_part = int_part[-3:]
        format_string = '0' * num_zeros_needed + int_part + '.' + decimal_part.ljust(13, '0')

    return format_string

def format_list_of_numbers(number_list):
    formatted_strings = []
    for number in number_list:
        formatted_string = generate_formatted_string(number)
        formatted_strings.append(formatted_string)
    return formatted_strings


ge = 0.10

def function(X,Y):
    Z = -np.cos(X) - np.cos(Y)- np.cos(X*Y) + ge * (X*X + Y*Y)
    return Z

def dervir_x(X,Y):
    return np.sin(X) +2* ge * (X) + np.sin(X*Y) * Y
    # return 
def dervir_y(X,Y):
    # return 2* ge*(Y) +  np.sin(X*Y) * X
    return 2* ge*(Y) +  np.sin(X*Y) * X + np.sin(Y)

Randoms = int(input())
# starting = tuple(map(int, starting_input.split(',')))

def iter(dual):
    X, Y = dual
    if Randoms == 1:
        lr = np.random.normal(0.005,0.05)
        lr_b = np.random.normal(0,0.00005)
        X -= lr * (dervir_x(X,Y)) + lr_b
        lry = np.random.normal(0.005,0.05)
        lr_by = np.random.normal(0, 0.00005)
        Y -= lry * (dervir_y(X,Y)) + lr_by
    else:
        lr = 0.1
        X -= lr * (dervir_x(X,Y))
        Y -= lr * (dervir_y(X,Y))

    return X,Y


data_points = deque(maxlen=1000)
X = np.linspace(-20, 20, 50)
Y = np.linspace(-20, 20, 50)
X, Y = np.meshgrid(X, Y)
Z = -np.cos(X) - np.cos(Y)- np.cos(X*Y) + ge * (X*X + Y*Y)
Z[(X >= 5) & (Y <= -10)] = np.nan

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.5)

scatter = ax.scatter([], [], [], c='r', marker='o', s=10,zorder=-1)
display(fig)
C = iter((-15,18))    
r = 0
iter1 = 0
trigger = False

for i in range(100000000):
    C = iter(C)
    X, Y = C
    Z = function(X,Y)

    data_points.append((X, Y, Z))

    x_values = [x for x, y, z in data_points]
    y_values = [y for x, y, z in data_points]
    z_values = [z for x, y, z in data_points]
    D = dervir_x(X,Y), dervir_y(X,Y)
    X = generate_formatted_string(X)
    Y = generate_formatted_string(Y)
    E = generate_formatted_string(Z)
    I = format_list_of_numbers(D)
    print_over(f"{X, Y, E, I , 'iter:',i}")

    if Z < -5:
        trigger = True
    
    if trigger == True:
        iter1 += 1
    
    if iter1 >= 3000:
        a = np.random.choice(np.arange(10,35))
        b = np.random.choice(np.arange(-35,-10))
        if r == 0:
            scatter.set_array([])

            scatter = ax.scatter([], [], [], c='b', marker='o', s=5,zorder=-1)
            r = 1
        else:
            scatter.set_array([])

            scatter = ax.scatter([], [], [], c='r', marker='o', s=5,zorder=-1)
            r = 0
        C = iter((a,b))  
        iter1 = 0 
        trigger = False

    if i % 1000 == 0:
                    
        clear_output(wait=True)
        display(fig)
        scatter._offsets3d = (x_values, y_values, z_values)
     

```
